# Credit Card Default Prediction

## Project Overview
This project aims to predict whether a credit card holder will default on their 
payment or not. Banks and financial institutions face huge losses when customers 
fail to repay their credit card bills. By using Machine Learning, we can identify 
risky customers in advance and take preventive action.

## Dataset Description
The dataset contains 2000 customer records with the following 5 variables:

- **Income** – The annual income of the customer
- **Age** – The age of the customer
- **Loan** – The loan amount taken by the customer
- **Loan to Income** – A calculated feature showing the ratio of loan to income (higher ratio = more risk)
- **Default** – Target variable: 1 means the customer defaulted, 0 means they did not

## Objective
To build a Machine Learning classification model that can accurately predict 
whether a customer will default on their credit card payment based on their 
financial information.

## Step 3: Define Target (y) and Features (X)

We separate the data into two parts:

- **X (Features)** – The input columns the model uses to make predictions:
  Income, Age, Loan, and Loan to Income ratio

- **y (Target)** – The column we want to predict:
  Default (1 = defaulted, 0 = did not default)

X is the information we give to the model, and y is the answer we want it to learn.

```python
# Step 3 : define target (y) and features (X)
```

```python
default.columns
```

```python
y = default['Default']
```

```python
X = default.drop(['Default'],axis=1)
```

## Step 4: Train Test Split

We split the dataset into two parts:
- **Training set (70%)** – The model learns patterns from this data
- **Testing set (30%)** – We use this to check how well the model performs on unseen data

We use **train_size=0.7** which means 70% goes to training and 30% goes to testing.
The **random_state=2529** ensures we get the same split every time we run the code.

This is important because if we test the model on the same data it learned from, 
we would not know how well it works on new, real-world data.

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Step 5: Select Model

We choose **Logistic Regression** as our Machine Learning model.

Even though the name says "regression", Logistic Regression is actually used for 
**classification problems** — predicting between two categories (Yes/No, 0/1, True/False).

It is a simple but powerful model that works well for predicting whether something 
will happen or not — in our case, whether a customer will default or not.

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

## Step 6: Train the Model

In this step, we train (or "fit") the model using the training data.

The model studies the relationship between the features (Income, Age, Loan, Loan to Income) 
and the target (Default). It finds patterns like:
- Customers with a high Loan-to-Income ratio are more likely to default
- Younger customers with large loans may be at higher risk

**model.intercept_** and **model.coef_** show the internal values the model learned during training.

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

## Step 7: Predict

Now we use the trained model to make predictions on the **test data** (data the model has never seen).

For each customer in the test set, the model outputs:
- **1** → The model predicts this customer WILL default
- **0** → The model predicts this customer will NOT default

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

## Step 8: Model Accuracy

Finally, we evaluate how well our model performed by comparing its predictions 
with the actual answers.

We use three evaluation tools:

1. **Confusion Matrix** – A table showing how many predictions were correct and incorrect:
   - True Positives: Correctly predicted defaults
   - True Negatives: Correctly predicted non-defaults
   - False Positives: Predicted default but customer did NOT actually default
   - False Negatives: Predicted no default but customer DID actually default

2. **Accuracy Score** – The percentage of correct predictions out of all predictions.
   For example, 0.85 means the model was correct 85% of the time.

3. **Classification Report** – A detailed report showing Precision, Recall, and F1-Score 
   for each category (0 and 1), giving a complete picture of model performance.

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
